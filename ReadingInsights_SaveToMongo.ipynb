# Clean the data usign Pandas library

```
import pandas as pd
import pandas as pd

# Load the Excel file into pandas (all sheets)
file_path = '/content/insights/Dr. Taste_apple_insights_AS.xlsx'  # Update with your actual file path
data = pd.read_excel(file_path, sheet_name=None, header=None)  # Load all sheets into a dictionary

cleaned_data = {}

# Process each sheet in the Excel file
for sheet_name, sheet_df in data.items():
    print(f"Processing sheet: {sheet_name}")

    # Step 1: Set the third row (index 2) as column names
    sheet_df.columns = sheet_df.iloc[2]  # Use third row (index 2) for column names

    # Step 2: Remove the first three rows (metadata and the old header)
    sheet_df = sheet_df[3:].reset_index(drop=True)  # Drop first 3 rows and reset index

    # Step 3: Clean the column names (strip spaces and convert to lowercase)
    sheet_df.columns = sheet_df.columns.str.strip().str.lower()  # Clean column names (optional)

    # Store cleaned data for this sheet
    cleaned_data[sheet_name] = sheet_df

    # Display the first few rows of the cleaned DataFrame (for inspection)
    print(sheet_df.head(10))  # Display the first 10 rows of cleaned data for this sheet
    print("\n")

# Now all sheets are processed and stored in `cleaned_data`
print("All sheets processed successfully!")
```

# Converto Items in the excel to JSON

```
import json

# Convert all sheets to JSON format
all_sheets_json = {}

for sheet_name, sheet_df in cleaned_data.items():
    # Convert DataFrame to list of dictionaries (JSON format)
    all_sheets_json[sheet_name] = sheet_df.to_dict(orient='records')  # 'records' format creates a list of dicts

# Save the JSON data to a file
output_json_path = '/content/insights/cleaned_all_sheets.json'  # Path to save the JSON file
with open(output_json_path, 'w', encoding='utf-8') as json_file:
    json.dump(all_sheets_json, json_file, ensure_ascii=False, indent=4)

print(f"All sheets have been saved as JSON: {output_json_path}")

```

# Import Sentence Transformer - Model:

```
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```
insights = [
    doc for doc in collection.find({"insight": {"$exists": True}}, {"_id": 0, "id": 1, "insight": 1})
]

embedding_list = []
for doc in insights:
    embedding = model.encode(doc["insight"])  # Generate embedding
    embedding_list.append({"id": doc["id"], "embedding": embedding})

print(f"Generated embeddings for {len(embedding_list)} documents.")
```

# Save the embeddings to Mongo DB by matching the unique ID

```
# Save embeddings back into the MongoDB collection
for item in embedding_list:
    collection.update_one(
        {"id": item["id"]},  # Match the document by its unique 'id'
        {"$set": {"embedding": item["embedding"].tolist()}}  # Save the embedding as a list
    )
print("Embeddings successfully saved to MongoDB!")
```

```
# Query embeddings and metadata from MongoDB
embedding_docs = list(collection.find(
    {"embedding": {"$exists": True}},  # Only fetch documents with embeddings
    {"_id": 0, "id": 1, "insight": 1, "embedding": 1}  # Select required fields
))

print(f"Retrieved {len(embedding_docs)} documents with embeddings.")
```

# Upload Embeddings to Chroma

```
import chromadb

# Initialize Chroma client
client = chromadb.Client()

# Retrieve the existing collection or create a new one
try:
    vector_collection = client.get_collection(name="insights_embeddings")
    print("Collection 'insights_embeddings' retrieved successfully.")
except ValueError:
    vector_collection = client.create_collection(name="insights_embeddings")
    print("Collection 'insights_embeddings' created successfully.")

# Add documents, embeddings, and metadata to the collection
for doc in embedding_docs:
    vector_collection.add(
        ids=[doc["id"]],               # Use the document's 'id' as the unique identifier
        documents=[doc["insight"]],    # Add the original insight text
        embeddings=[doc["embedding"]], # Add the embedding
        metadatas={"id": doc["id"]}    # Add metadata like document ID
    )
print("Embeddings successfully uploaded to Chroma!")
```

```
# Example query
query = "How does inflation affect consumer behavior?"
query_embedding = model.encode(query)  # Generate embedding for the query

# Query the vector database for top 3 similar documents
results = vector_collection.query(
    query_embeddings=[query_embedding],
    n_results=3  # Number of most similar documents to retrieve
)

# Print the top results
print("Top results:")
for idx, doc in enumerate(results['documents'][0]):  # Iterate over the first list of documents
    metadata = results['metadatas'][0][idx]  # Access metadata for the current document
    print(f"Document ID: {metadata['id']}, Text: {doc}")
```
